```python
#separating dependent and independent variables
X = Loan1_df.drop('Personal Loan', axis=1)
Y = Loan1_df[['Personal Loan']]
```

```python
# Splitting the data
from sklearn.model_selection import train_test_split
import statsmodels.api as sm

X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size = 0.3, random_state = 42 )

logit = sm.Logit( y_train, sm.add_constant( X_train ) )

lg = logit.fit()

lg.summary2()
```

```python
from sklearn import metrics

from sklearn.linear_model import LogisticRegression


model1 = LogisticRegression(solver="liblinear")
model1.fit(X_train, y_train)

y_predict = model1.predict(X_test)


coef_df = pd.DataFrame(model1.coef_)
coef_df['intercept'] = model1.intercept_
print(coef_df)
```

```python
model1_score = model1.score(X_test, y_test)
print(model1_score)
```

```python
confum=metrics.confusion_matrix(y_test, y_predict, labels=[1, 0])

df_confum = pd.DataFrame(confum, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))

sns.heatmap(df_confum, annot=True)
```

```python
from sklearn.metrics import confusion_matrix, recall_score, precision_score, f1_score, roc_auc_score,accuracy_score
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=42)
logreg.fit(X_train, y_train) 
y_predict = logreg.predict(X_test)

print("Trainig accuracy",logreg.score(X_train,y_train))  
print()
print("Testing accuracy",logreg.score(X_test, y_test))
print()
print("Recall:",recall_score(y_test,y_predict))
print()
print("Precision:",precision_score(y_test,y_predict))
print()
print("F1 Score:",f1_score(y_test,y_predict))
print()
print("Roc Auc Score:",roc_auc_score(y_test,y_predict))
```

```python
#checking different values
train_score=[]
test_score=[]
solver = ['newton-cg','lbfgs','liblinear','sag','saga']
for i in solver:
    model2 = LogisticRegression(random_state=42,penalty='l2', C = 0.75,solver=i)  
    model2.fit(X_train, y_train) 
    y_predict = model2.predict(X_test)     
    train_score.append(round(model2.score(X_train, y_train),3))
    test_score.append(round(model2.score(X_test, y_test),3))
    
print(solver)
print()
print(train_score)
print()
print(test_score)
```

```python
#Highest accuracy is with using 'l2' with 'newton-cg, Model performance will be affected by using 'l1'with Liblinear or saga, or using 'l2' with lbfgs and sag
```

```python
# Checking for C values
train_score=[]                                 
test_score=[]
C = [0.01,0.1,0.25,0.5,0.75,1]
for i in C:
    model3 = LogisticRegression(random_state=42,penalty='l1', solver='liblinear',class_weight='balanced', C=i)  # changing values of C
    model3.fit(X_train, y_train) 
    y_predict = model3.predict(X_test)     
    train_score.append(round(model3.score(X_train,y_train),3)) # appending training accuracy in a blank list for every run of the loop
    test_score.append(round(model3.score(X_test, y_test),3))   # appending testing accuracy in a blank list for every run of the loop
    
print(C)
print()
print(train_score)
print()
print(test_score)
```

```python
# the best prediction is gotten at C=0.5, 0.75 and 1
```

```python
#True Positive (observed=1,predicted=1):
#Predicted that personal loan will be accepted by the customer who was eligibile for loan

#False Positive (observed=0,predicted=1):
#Predicted that personal loan will be accepted by the customer who was not eligibile for loan

#True Negative (observed=0,predicted=0):
#Predicted that personal loan will NOT be accepted by the customer who was NOT eligibile for loan

#False Negative (observed=1,predicted=0):
#Predicted that personal loan will NOT be accepted by the customer who was eligibile for loan

#The bank wants to give personal loans to all eligible customers from the problem statement.This way the number of customers who would accept the personal loan program would be maximized and they would see an increase in the loan issued. However, the bank only wants to give the loans to eligible customers. The important factor to consider is the False Negatives where we predict wrongly and miss out on customers who would be eligible to accept the personal loan. Precision is the important metric.
#In case of False negative which is most key in this problem, the bank will lose the TARGET customers they desire to capture or convert from being liability customers.
```
